```python
import numpy
import json 
import pandas as pd 
import requests
import time
```

# Importing Data from Json File

```python
response = requests.get('https://data.covid19india.org/v4/min/data.min.json')
json_file= response.json()
```

```python
import pandas as pd

# Assuming json_file is the variable containing your JSON data
data_struc = []

for state, state_data in json_file.items():
    if 'districts' in state_data:
        for district in state_data['districts']:
            data_struc.append({'State': state, 'District': district})

district_main_table = pd.DataFrame(data_struc)
district_main_table
```

```python
state_districts = {}

for state, state_data in json_file.items():
    if 'districts' in state_data:
        state_districts[state] = list(state_data['districts'].keys())

state_districts
```

# Delta Data

```python
import numpy as np

# Initialize the dictionary to store delta values
district_delta_table = {'confirmed': [], 'deceased': [], 'recovered': [], 'vaccinated1': [], 'vaccinated2': [], 'tested': []}

# Loop through each metric (column) in the district_delta_table
for column in district_delta_table.keys():
    # Use list comprehension to extract delta values for each metric from the JSON data
    delta_values = [json_file[state]['districts'].get(district, {}).get('delta', {}).get(column, 0) for state, districts in state_districts.items() for district in districts]
    # Append the delta values to the corresponding list in the district_delta_table
    district_delta_table[column] = delta_values

# Create DataFrame from district_delta_table
district_delta_table_structure = pd.DataFrame(district_delta_table)

# Display the DataFrame
district_delta_table_structure
```

# Delta7 Table 

```python
# Initialize the dictionary to store delta7 values
district_delta7_table = {'confirmed': [], 'deceased': [], 'recovered': [], 'vaccinated1': [], 'vaccinated2': [], 'tested': []}

# Loop through each metric (column) in the district_delta7_table
for column in district_delta7_table.keys():
    # Use list comprehension to extract delta7 values for each metric from the JSON data
    delta7_values = [json_file[state]['districts'].get(district, {}).get('delta7', {}).get(column, 0) for state, districts in state_districts.items() for district in districts]
    # Append the delta7 values to the corresponding list in the district_delta7_table
    district_delta7_table[column] = delta7_values

# Create DataFrame from district_delta7_table
district_delta7_table_structure = pd.DataFrame(district_delta7_table)

# Display the DataFrame
district_delta7_table_structure
```

# Meta Data 

```python
# Initialize the dictionary to store metadata values
district_meta_table = {'population': []}

# Loop through each metadata field (column) in the district_meta_table
for column in district_meta_table.keys():
    # Use list comprehension to extract metadata values for each field from the JSON data
    meta_values = [json_file[state]['districts'].get(district, {}).get('meta', {}).get(column, 0) for state, districts in state_districts.items() for district in districts]
    # Append the metadata values to the corresponding list in the district_meta_table
    district_meta_table[column] = meta_values

# Create DataFrame from district_meta_table
district_meta_table_structure = pd.DataFrame(district_meta_table)

# Display the DataFrame
district_meta_table_structure
```

# Dictionaries to store Data

```python
# Initialize dictionary to store total values
district_total_table = {'confirmed': [], 'deceased': [], 'recovered': [], 'vaccinated1': [], 'vaccinated2': [], 'tested': []}

# Loop through each metric
for column in district_total_table.keys():
    # Extract total values for each metric from the JSON data
    total_values = [json_file[state]['districts'].get(district, {}).get('total', {}).get(column, 0) for state, districts in state_districts.items() for district in districts]
    # Append total values to the corresponding list
    district_total_table[column] = total_values

# Create DataFrame
district_total_table_structure = pd.DataFrame(district_total_table)

# Display DataFrame
district_total_table_structure
```
